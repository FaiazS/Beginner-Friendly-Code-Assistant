```python
from litellm import completion

from typing import List, Dict

def generate_response(messages: List[Dict]) -> str:

 llm_response = completion(

                          model = "groq/llama-3.3-70b-versatile",

                          messages = messages,

                          max_tokens = 2500
  )

 return llm_response.choices[0].message.content

code_query = input("Hey there, kindly input your requirement for your coding project and I will be most glad to assist you in the same")


messages = [

               {

                 "role" : "system",

                 "content" : "You are an exceptionally skilled Software Engineering Virtual Assistant who codes both in the functional programming and object orientated style, your primary role is to assist software engineers of any level with any coding or debugging requirement that they may have, always assume that they are just starting out in Software Engineering, thus always provide solutions with clear and easy comments and line by line explanation and last but not the least, speak out loud and explain your thought process of why you feel that the solution you provided will work and handles trade offs efficiently and optimized for a specific usecase."

               },

              {

                 "role" : "user",

                 "content" : code_query


                 }

             ]

codebot_response = generate_response(messages)

print(codebot_response)
```

```python
messages = [

       {"role" : "assistant" , "content" : codebot_response},

       {"role" : "user" , "content" : "Update the implementation provided to include beginner-friendly documentation so that future developers may also understand the solution that you provided."}

        ]
```

```python
codebot_response = generate_response(messages)

print(codebot_response)
```
